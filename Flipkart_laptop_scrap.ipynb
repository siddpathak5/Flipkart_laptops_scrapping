```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.flipkart.com/laptops-store?fm=neo%2Fmerchandising&iid=M_c913497e-a8b5-412e-8c7e-e70e33968f4a_1_372UD5BXDFYS_MC.WB1CFS0X26D1&otracker=hp_rich_navigation_2_1.navigationCard.RICH_NAVIGATION_Electronics~Laptop%2Band%2BDesktop~Laptops_WB1CFS0X26D1&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_2_L2_view-all&cid=WB1CFS0X26D1')
```

```python
soup=BeautifulSoup(driver.page_source,'html.parser')
```

```python
soup
```

```python
names = "wjcEIp"
price = "Nx9bqj"
rating = "XQDdHH"
```

```python
laptop_name = soup.find_all('a', class_=names)
laptop_price = soup.find_all('div', class_=price)
laptop_rating = soup.find_all('div', class_=rating)
```

```python
laptop_name
```

```python
laptop_name_list = []
for i in laptop_name:
    laptop_name_list.append(i.text)
```

```python
laptop_name_list
```

```python
laptop_price
```

```python
laptop_price_list = []
for i in laptop_price:
    laptop_price_list.append(i.text)
```

```python
laptop_price_list
```

```python
laptop_rating
```

```python
laptop_rating_list = []
for i in laptop_rating:
    laptop_rating_list.append(i.text)
```

```python
laptop_rating_list
```

```python
driver.quit()
```

```python
data = {
    'Name': laptop_name_list,
    'Price': laptop_price_list,
    'Rating': laptop_rating_list
}
df = pd.DataFrame(data)
```
